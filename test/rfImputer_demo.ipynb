```python
import pandas as pd
import sys
sys.path.insert(1, '../')
from rfImputer import rfImputer
from sklearn.preprocessing import Imputer
from pprint import pprint

cols = ['Proof_cut', 'Viscosity', 'Caliper', 'Ink_temperature', 'Humifity',
        'Roughness', 'Blade_pressure', 'Varnish_pct', 'Press_speed', 'Ink_pct',
        'Solvent_pct', 'Esa_voltage', 'ESA_amperage', 'Wax', 'Hardener',
        'Roller_durometer', 'Density', 'Anode_ratio', 'Chrome_content', 'bands']
df = pd.read_csv('bands.csv')
df.columns = cols

df.replace('?', 'NaN', inplace = True)
df.replace('band', '1', inplace = True)
df.replace('noband', '0', inplace = True)

for col in df:
    if col == 'bands':
        continue
    df[col] = df[col].astype(float)
```

```python
df
```

```python
imputer = Imputer(missing_values = 'NaN', strategy = 'mean', copy = False)
imputer.fit(df)
df_new = imputer.transform(df)
pd.DataFrame(df_new).head(20)
```

```python
# Instantiate Imputer with data
imp_df = rfImputer(df, is_classification = ['Density', 'Chrome_content'],
                  is_regression = ['Anode_ratio'],
                  excl_impute = ['Caliper', 'Ink_pct'], # or incl_impute = 
                  incl_predict = ['Press_speed', 'Solvent_pct', 'Chrome_content']) # or excl_predict = 
```
